```python
import numpy as np

a1 = np.arange(30).reshape(3,5,2)
print(a1, a1.shape)
```

```python
a2 =np.sum(a1, axis = 0)
print(a2, a2.shape)  # 0 + 10 + 20
```

```python
a3 =np.sum(a1, axis = 2)
print(a3, a3.shape)
```

```python
a4 = np.sum(a1, axis = 1)
print(a4, a4.shape)
```
