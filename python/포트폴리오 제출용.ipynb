```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family']=[kfont]+current_font_list



pd_data= pd.read_csv('소비자물가지수_2020100__20230313160626.csv', encoding = 'cp949')
print(pd_data)
print(pd_data.dtypes)

pd_data['시점'] = pd_data['시점'].astype(int)
print(pd_data.dtypes)
```

```python
_,axe = plt.subplots()
axe.plot(pd_data['부산광역시'])
```

```python
fig = plt.figure(figsize=(6.4*2,4.8))
left = fig.add_subplot(1,2,1)
left.scatter(pd_data['시점'],pd_data['전국'])
```

```python
sns.lineplot(data=pd_data, x="시점", y="전국")
```

```python
pd_data.plot(x='시점', y = '전국')
```
