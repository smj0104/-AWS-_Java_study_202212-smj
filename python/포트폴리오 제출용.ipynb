    * 시나리오/스토리보드 개발 : 코로나가 진행될때 조금씩 상승하다 코로나 말기에 급격한 상승을 보임
### 데이터 시각화 명세서
    * 시각화 표현의 형태,규칙,패턴,상호작용 : 년도별,지방별로 pandas와 seaborn을 사용하여 표현
### 시각화 명세서 기반 설계
    * 시각화 UI 설계 : 대한민국을 지방으로 분리, lineplot을 사용하여 지방별 물가 상승률과 물가가 크게 오른 시점을 조사
    

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family']=[kfont]+current_font_list



pd_data= pd.read_csv('소비자물가지수_2020100__20230313160626.csv', encoding = 'cp949')
print(pd_data)
print(pd_data.dtypes)

pd_data['시점'] = pd_data['시점'].astype(int)
print(pd_data.dtypes)
```

```python
sns.set_theme(style="darkgrid")
pd_data.head()
```

```python
axe = sns.lineplot(data=pd_data, x="시점", y="서울특별시",label = '서울').set(title='서울지역 물가 상승률')
axe = sns.lineplot(data=pd_data, x="시점", y="인천광역시",label = '인천')
axe = sns.lineplot(data=pd_data, x="시점", y="세종특별자치시",label = '세종')
axe = sns.lineplot(data=pd_data, x="시점", y="경기도",label = '경기')
axe = sns.lineplot(data=pd_data, x="시점", y="강원도",label = '강원')

axe.set_xlabel('소비자물가지수')
axe.set_ylabel('물가상승시점')
axe.legend(loc='upper left')
```

```python
axe = sns.lineplot(data=pd_data, x="시점", y="부산광역시",label = '부산').set(title='경상지역 물가 상승률')
axe = sns.lineplot(data=pd_data, x="시점", y="대구광역시", label = '대구')
axe = sns.lineplot(data=pd_data, x="시점", y="울산광역시", label = '울산')
axe = sns.lineplot(data=pd_data, x="시점", y="경상남도", label = '경상')
axe = sns.lineplot(data=pd_data, x="시점", y="경상북도" ,label = '경북')

axe.set_xlabel('소비자물가지수')
axe.set_ylabel('물가상승시점')
axe.legend(loc='upper left')
```

```python
axe = sns.lineplot(data=pd_data, x="시점", y="광주광역시",label = '광주').set(title='호남지역 물가 상승률')
axe = sns.lineplot(data=pd_data, x="시점", y="전라남도", label = '전남')
axe = sns.lineplot(data=pd_data, x="시점", y="전라북도" ,label = '전북')

axe.set_xlabel('소비자물가지수')
axe.set_ylabel('물가상승시점')
axe.legend(loc='upper left')
```

```python
axe = sns.lineplot(data=pd_data, x="시점", y="세종특별자치시",label = '세종').set(title='대전지역 물가 상승률')
axe = sns.lineplot(data=pd_data, x="시점", y="대전광역시", label = '대전')
axe = sns.lineplot(data=pd_data, x="시점", y="충청남도", label = '충남')
axe = sns.lineplot(data=pd_data, x="시점", y="충청북도", label = '충북')

axe.set_xlabel('소비자물가지수')
axe.set_ylabel('물가상승시점')
axe.legend(loc='upper left')
```

### 물가상승률에 대해
    * 일반적으로 서울이 물가가 가장 높고 물가상승률이 높다고 알려져있지만,놀랍게도 모든 지방이 서울의
    물가를 따라가는 모습을 보여주었다. 오히려 경기도는 서울보다 물가상승이 높기도하였고 경북지역은
    가장 높은 물가상승률을 보여주기도 하였다.
