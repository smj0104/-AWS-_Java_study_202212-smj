```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']  ## matplotlib이 내부적으로 가진 변숫값 표현

font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family']=[kfont]+current_font_list


pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_20230303160532.csv', encoding = 'cp949')
print(pd_data.head())
```

```python
fig,axes = plt.subplots(2,4, figsize = (6.4*4,4.8*2))
col_names = pd_data.columns[-2:]

for i,n in enumerate(pd_data['데이터직무별(1)'].unique()):
    filter1 = pd_data['데이터직무별(1)']==n
    xs = pd_data.loc[filter1, '시점']
    ys = pd_data.loc[filter1, col_names]
    ax.set_title(n)
    ax.plot(xs.astype('str'),ys.iloc[:0], 'go--')
    ax.plot(xs.astype('str'),ys.iloc[:1], 'ro--')
    
    #print(xs.dtype)
```

```python
filter1 = pd_data['데이터직무별(1)']=='데이터 아키텍트'
col_names = pd_data.columns[-2:]
xs = pd_data.loc[filter1, '시점']
ys = pd_data.loc[filter1, col_names]
_,axe = plt.subplots()
axe.plot(xs, ys)
```
