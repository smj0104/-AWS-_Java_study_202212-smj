### 전국횡단보도표준데이터

- 횡단보도연장, 녹색신호시간 : 상관도
- 자전거 횡단도 겸용 비율(전체 대비), 차로수별 자전거 횡단도 카운트/비율
- 차로수별 보행자 신호등 유무 카운트/비율
- 차로수별 음향신호기 설치의 유무 카운트/비율

### 화면 출력
1.상관도 수치
2.비율 수치
3~5 카운트,비율


```python
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i].strip())
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

raw_data=[]
with open('전국횡단보도표준데이터.csv') as f:
    for line in f:
        raw_data.append(line[:-1])
        
first_data= (my_split(raw_data[0]))
road_length = first_data.index('횡단보도연장')

road_length_list=[]
for i in range(1, len(raw_data)):
    a = my_split(raw_data[i])
    road_length_list.append(a[road_length])
    
print(road_length_list)
```

```python
light_length = first_data.index('녹색신호시간')

greenlight_length=[]
for i in range(1, len(raw_data)):
    a = my_split(raw_data[i])
    greenlight_length.append(a[light_length])
    
print(greenlight_length)
```

```python
bicycle_cross = first_data.index('자전거횡단도겸용여부')

bicycle_cross_list=[]
for i in range(1, len(raw_data)):
    a = my_split(raw_data[i])
    bicycle_cross_list.append(a[bicycle_cross])
#print(bicycle_cross_list)


total_count = len(bicycle_cross_list)

print('Y:', bicycle_cross_list.count('Y'), ' N:', bicycle_cross_list.count('N'), '없음:', bicycle_cross_list.count('') )
print('Y: ', (bicycle_cross_list.count('Y') / total_count)*100, '%')
```
