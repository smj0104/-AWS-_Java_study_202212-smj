```python
import numpy

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
data=[]
with open('전국공인중개사사무소표준데이터.CSV') as a:
    for line in a:
        data.append(line)
        
#print(my_split(data))

agent_data=(my_split(data[0]))
estate_agent_data = agent_data.index('개업공인중개사종별구분')
percentage=[]

for i in range(1, len(data)):
    b = my_split(data[i])
    percentage.append(b[estate_agent_data])
    
#print(percentage)

list(set(percentage))
print('공인중개사 : ',percentage.count('공인중개사'))
print('법인 : ',percentage.count('법인'))
print('개업공인중개사 : ',percentage.count('개업공인중개사'))

x = percentage.count('공인중개사')
y = percentage.count('법인')
z = percentage.count('개업공인중개사')

x_percent = (x/29369)*100
y_percent = (y/29369)*100
z_percent = (z/29369)*100

print('공인중개사 : ', x_percent,'%')
print('법인 : ', y_percent, '%')
print('개업공인중개사 : ', z_percent, '%')
```
