```python
# '시도명' 별 hydrant, parking count

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_water = pd.read_csv('전국소방용수시설표준데이터.csv', encoding = 'cp949')
pd_car = pd.read_csv('전국소방자동차전용구역표준데이터.csv', encoding = 'cp949')


g2 = pd_water.groupby('시도명')
g3 = pd_car.groupby('시도명')

#df1 = pd_water[['시도명','소재지지번주소','지자체명']]
#df2 = pd_car[['시도명','소재지지번주소','시도명']]

col_selection = ['시도명', '소재지도로명주소', '위도', '경도']
df1 = pd_water[col_selection]
df2 = pd_car[col_selection]

df_combined = pd.concat((df1,df2), keys =['hydrant', 'parking'])
print(df_combined.shape)  #(53492, 4)
df_combined.reset_index(inplace = True)
df_combined.drop('level_1', axis=1,inplace=True)

cname_list = list(df_combined.columns)
cname_list[0]='종류'
df_combined.columns = cname_list

df_combined





```
