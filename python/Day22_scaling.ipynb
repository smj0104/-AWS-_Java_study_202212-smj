```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
pd_data = pd.read_csv(url, sep = '\t')
print(pd_data.head())
print(pd_data.shape)
```

```python
_, axe = plt.subplots()
#axe.plot(pd_data['BMI'])
axe.plot(pd_data['BP'])
```

```python
# normalization   : (x - x_min) / (x_max-x_min)
pd_data['BMI'].describe()
bmi_min = pd_data['BMI'].min()
bmi_max = pd_data['BMI'].max()
pd_data['BMI_norm'] = pd_data['BMI'].apply(lambda x: (x-bmi_min)/(bmi_max-bmi_min))
```

```python
print(pd_data)
```

```python
pd_data['BP'].describe()
bmi_min = pd_data['BP'].min()
bmi_max = pd_data['BP'].max()
pd_data['BP_norm'] = pd_data['BP'].apply(lambda x: (x-bmi_min)/(bmi_max-bmi_min))

_, axe = plt.subplots()
axe.plot(pd_data['BMI_norm'])
axe.plot(pd_data['BP_norm'])

```

```python
# standardization : (x - x_mean)/x_std
# -   () .

pd_data['BMI'].describe()
bmi_mean = pd_data['BMI'].mean()
bmi_std = pd_data['BMI'].std()
pd_data['BMI_std'] = pd_data['BMI'].apply(lambda x: (x-bmi_mean)/bmi_std)

#print(pd_data['BMI_std'].describe())
```
