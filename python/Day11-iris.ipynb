```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

with open('iris_ndarray.pickle','rb') as f:
    iris_np = pickle.load(f)
print(iris_np)
sepal_length = iris_np[:,0]
sepal_width = iris_np[:,1]
sepal_ratio = sepal_length / sepal_width
#print(sepal_ratio)


petal_length = iris_np[:,2]
petal_width = iris_np[:,3]
petal_ratio = petal_length / petal_width
#rint(petal_ratio)

print(sepal_ratio.shape)
sepal_ratio_2d = np.expand_dims(sepal_ratio, 1)  #디멘션 하나 추가함 1의 자리에 0을 주면 0번의 자리에 디멘션이 들어감(0, 150)
print(sepal_ratio_2d.shape)

print(petal_ratio.shape)
petal_ratio_2d = petal_ratio[:, np.newaxis] #어느 위치에 넣을 것인지 조절가능
print(petal_ratio_2d.shape)


concate_result = np.concatenate((sepal_ratio_2d, petal_ratio_2d), axis=1)  #축 하나 더하기
print(concate_result.shape)
print(concate_result[:10])
#0번 나누기 1번  을 모아놓은 리스트
#3번 나누기 4번  을 모아놓은 리스트

#print(np.divide.(iris_np))
```
