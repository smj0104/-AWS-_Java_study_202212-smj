```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tips = sns.load_dataset('tips')
print(df_tips)
```

```python
# hue, style, size
sns.relplot(data=df_tips, x = 'total_bill', y ='tip', hue ='sex', style ='smoker' )
#row, col
sns.relplot(data=df_tips, x = 'total_bill', y ='tip', col ='sex', style ='smoker' )
```
