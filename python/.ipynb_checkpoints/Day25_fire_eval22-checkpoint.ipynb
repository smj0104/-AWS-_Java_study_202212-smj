```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

current_font_list = matplotlib.rcParams['font.family']  ## matplotlib이 내부적으로 가진 변숫값 표현

font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family']=[kfont]+current_font_list


pd_data = pd.read_csv('시도별_화재발생_현황_총괄__22.csv', encoding = 'cp949')
print(pd_data.head())
```

```python
df_tips = sns.load_dataset('tips')
sns.displot(data=df_tips, x='total_bill',hue = 'time', kind ='ecdf')
sns.displot(data=df_tips, x='time')  # categorial var.
#sns.catplot(data=df_tips, x='time')
```
