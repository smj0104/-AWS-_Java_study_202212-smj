```python
print(np.pi)
print(np.exp(5))
print(np.sqrt(25))
#변수
mu = 0  # 평균
sig = 3 # 표준편차, 분산의 제곱근
x = np.linspace(-5,5,100)
y = 1/sig*np.sqrt(2 * np.pi) * np.exp(-((x-mu)*x-mu))/ (2*sig*sig)
#y = y + np.exp(-((x_mu)**2)/(2*sig*sig))
_,axe = plt.subplots()
axe.plot(x,y)
```

```python
import math

mu = 0
sig = 1
x = np.linspace(-5,5,100)
y = []
for v in x:
    y.append(0.5*(1+math.erf((v-mu)/(sig*np.sqrt(2)))))
    
    
_,axe = plt.subplots()
axe.plot(x,y)
```
