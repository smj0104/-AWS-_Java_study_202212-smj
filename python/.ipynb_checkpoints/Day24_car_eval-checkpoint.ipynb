```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data'
pd_data = pd.read_csv(url, header = None)
pd_data.columns = ['구매가','유지비용','문갯수','탑승인원','수납공간','안전도','만족도']
print(pd_data.head())
```

```python
_,axe = plt.subplots()
axe.plot(pd_data['구매가'])
```

```python
_,axe = plt.subplots()
axe.pie(pd_data['수납공간'].value_counts())
```

```python
# 구매가가 만족도에 미치는 영향
    # figure를 좌우로 나누어, 왼쪽은 구매가vs맍녹도 산점도, 오른쪽은 아래위로 분할,
    #위쪽은 구매가 vhigh중 만족도 count, 아래쪽은 구매가 low중 만족도별 count(bar)
    
# 수납공간이 크면 탑승인원도 많은가.
```

```python
fig = plt.figure(figsize=(6.4*2,4.8))
left = fig.add_subplot(1,2,1)
left.scatter(pd_data['구매가'],pd_data['만족도'], s =100)

right_top = fig.add_subplot(2,2,2)
filter1 = pd_data['구매가']=='vhigh'
s1 = pd_data.loc[filter1, '만족도'].value_counts()
s2 = pd.Series(list(s1.values)+[0,0], index = ['unacc', 'acc', 'good', 'vgood'])
right_top.bar(s1.index, s1)
#print(s2)
#print(s1.values)

right_bot = fig.add_subplot(2,2,4)
filter1 = pd_data['구매가']=='low'
s1 = pd_data.loc[filter1, '만족도'].value_counts()
right_bot.bar(s1.index, s1)

from pandas.api.types import CategoricalDtype

price_category = CategoricalDtype(categories=["low", "med", "high", 'vhigh'], ordered=True)
eval_catetgory = CategoricalDtype(categories=["unacc", "acc", "good", 'vgood'], ordered=True)

g1 = pd_data.groupby(['구매가', '만족도'])
s3 = g1['안전도'].count()
s3 = s3.reset_index()
s3['구매가'] = s3['구매가'].astype(price_category)
s3.sort_values('구매가', inplace = True)
print(s3)
left.scatter(s3['구매가'], s3['만족도'],s=s3['안전도'])
```
