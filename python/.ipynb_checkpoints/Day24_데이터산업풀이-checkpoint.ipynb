```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']  ## matplotlib이 내부적으로 가진 변숫값 표현

font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family']=[kfont]+current_font_list


pd_data = pd.read_csv('데이터산업_시장규모_20230306153117.csv', encoding = 'cp949')
print(pd_data)
```

```python
fig = plt.figure(layout = 'tight')
line_axe = fig.add_subplot(2,1,1)
col_names = pd_data.columns
line_params = ['ro-','g^--', 'bv:']
line_axe.set(title = '2015~2021 산업별 규모', ylabel = '억원')
#line_axe.setr_ylable('억원')
#line_axe.set(title='2015~20')

for i in range(1,4):
    #line_axe.plot(pd_data['시점'], pd_data[col_names[i]], line_params[i-1])
    line_axe.plot('시점', col_names[i], line_params[i-1], data = pd_data)

pie2015_axe = fig.add_subplot(2,2,3)
pie2015_axe.set_title('2015년 구성비', y=-0.1)
pie2015_axe.pie(pd_data.iloc[0,1:])

pie2021_axe = fig.add_subplot(2,2,4)
pie2021_axe.set_title('2021년 구성비', y=-0.1)
pie2021_axe.pie(pd_data.iloc[-1,1:],labels = list('abc'), autopct='%.2f')
```
