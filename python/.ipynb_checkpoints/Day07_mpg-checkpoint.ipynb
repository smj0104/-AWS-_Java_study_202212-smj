```python
import math

def mean(num_list):
    return sum(num_list)/len(num_list)

def mean2(num_list):
    sum_valu=()
    for i in num_list:
        sum_value += i
    return sum_value / len(num_list)

def median(num_list):  #중앙값
    num_list.sort()
    if len(num_list)%2==1:
        i = (len(num_list))//2
        return num_list[i]
    else:
        i = len(num_list)//2
        return (num_list[i] + num_list[i-1])/2
    
    
def dev(num_list): #편차
    m = mean(num_list)
    return [x-m for x in num_list]


def var(num_list): #분산
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)


def covar(list_1,list_2): #공분산
    n = len(list_1)
    list_1_dev = dev(list_1)
    list_2_dev = dev(list_2)
    return sum(x * y for x, y in zip(list_1_dev, list_2_dev))/(n-1)


def corel(list_1, list_2): # 상관도 (-1 ~ 1)  마이너스라도 1에 가까우면 상관도 높음
    return covar(list_1, list_2)/(stdev(list_1)*stdev(list_2))


def stdev(num_list): #표준편차  (분산에 루트)
    return math.sqrt(var(num_list))
```

```python
import requests

x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data')
```

```python
car_data = []
for line in x.text.split('\n'):
    car_data.append(line)
    
print(car_data)
```

```python
pass1 = x.text.split('\n')  #by_line
#print(len(pass1))

pass2 = []   #각 샘플별 숫자부분, 차명 부분 분리, split('\t')
for line in pass1:
   # print(line.split('\t'))
    pass2.append(line.split('\t'))
    
pass3 = []  # 숫자 부분 분리 .split()
for e in pass2:
    try:
        y = e[0].split()
        y.append(e[1])
        pass3.append(y)
    except:
        pass

print(pass3)

#y = pass2[0][0].split()
#y.append(pass2[0][1])
#print(y)

#pass2[0][1]
```

```python
#float으로 형변환, 차명 "" 제거하기'''
for e in pass3:
    for i,n in enumerate(e[:-1]):
        try:
            e[i] = float(n)
        except:
            print(n)
            e[i] = 0.0
            
    e[-1] = e[-1].strip('"')
    
    
    
    
    
    
#for e in pass3[0][:-1]:
    #print(float(e))
   

print(e)
```

```python
#rint(pass3)
mpg_list=[]  #연비
disp_list=[] #배기량
for e in pass3:
    mpg_list.append(e[0])
    disp_list.append(e[2])
    
print(e)
```

```python
#연비 : 평균, 표준편차
print('평균:', mean(mpg_list))
print('표준편차:', stdev(mpg_list))

#배기량: 평균, 표준편차
print('평균:', mean(disp_list))
print('표준편차:', stdev(disp_list))

```

```python
corel(mpg_list, disp_list)
```

```python
#차체중량(인덱스4) vs.연비
weight_list = [e[4] for e in pass3]
print('상관도:', corel(mpg_list, weight_list))
```
