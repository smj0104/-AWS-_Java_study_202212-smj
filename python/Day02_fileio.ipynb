```python
data_list = []
with open('부산광역시_노인복지관 현황_20230101.csv') as f: #csv comma seperated value  인코딩 필요할시
                                                            # ,encoding = 'utf-8' 혹은 'cp949' 넣어주기
    for line in f:
        data_list.append(line.split(','))
        
    
print(data_list)


#f.close() # open한 파일은 반드시 close
```

```python
gu_list=[]
for i in data_list:
    address = i[2]
    print(gu_list)
    address_list = address.split()
    gu_list.append(address_list)
#print(gu_list)
#print(gu_list.index('다대포'))
#gu_list[21] = '사하구'
#print(gu_list)
```

```python
gu_set = set(gu_list)
d_gu_list = []
d_gu_count= []
for i in gu_set:
    #print(i,gu_list.count(i))
    d_gu_list.append(i)
    d_gu_count.append(gu_list.count(i))
print(d_gu_list)
print(d_gu_count)
    
```
