```python
# 빈도수 dictionary 반환
# 1. raw_list를 set으로 변환
# 2. 셋을 순환하면서, 각각 몇개씩 raw_list에 있는지 카운트
# 3. 1,2,에서 얻은 key, value 값으로 dictionary 만들어 return.

list_key = []
list_value = []
def unique_count(raw_list, sort = False):
    raw_set = set(raw_list)
    for i in raw_set:
        list_key.append(i)
        list_value.append(raw_list.count(i))
        
    ret_dict = dict(zip(list_key,list_value))
    if sort == True:
        ret_dict = dict(sorted(ret_dict.items(), key= lambda x:x[1]))   
    return ret_dict
        #lambda 풀어서 def sort_key():
        #                  return x[1]
        #print(i, raw_list.count(i))
        
print(i)
        

    

unique_count(name_list)
```
