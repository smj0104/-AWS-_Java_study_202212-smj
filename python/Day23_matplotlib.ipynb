```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

#print(xs)
#print(ys)

figure, axe = plt.subplots()
figure.set_facecolor('blue')
figure.suptitle('Figure1')
figure.supxlabel('Figure1_x')
figure.supylabel('Figure1_y')
figure.text(0.5, 0.5 ,'abc fk')
figure.set_size_inches(15, 4.8)

axe.plot(xs, ys)
axe.set_title('my_axe_title')
axe.set_xlabel('abc x')
axe.set_ylabel('nok')
axe.set_xlim(0,20)
axe.set_ylim(0,1)
axe.set_xticks(np.linspace(0,30,20))

print(type(figure))
print(type(axe))
```
